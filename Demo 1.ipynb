```python
import pandas as pd
import simplejson
import matplotlib.pyplot as plt
import matplotlib.dates as md
import os
from monclient import client

%matplotlib inline

plt.rcParams.update({'font.size': 8})

endpoint = 'http://192.168.10.4:8080/v2.0'
api_version = '2_0'
kwargs = {
    'username': '',
    'password': '',
    'token': '82510970543135',
    'tenant_id': '82510970543135',
    'tenant_name': '',
    'auth_url': '',
    'service_type': '',
    'endpoint_type': '',
    'insecure': False,
    'runlocal': True
}

mon_client = client.Client(api_version, endpoint, **kwargs)
metrics = mon_client.metrics
```

```python
def measurements_to_dfs(measurements):
    dfs = []
    for m in measurements:
        df = pd.DataFrame(m['measurements'], columns=m['columns'])
        dfs.append(df)
    return dfs

def get_max_value(measurements):
    '''Get the max value of all measurements'''
    max_value = 0
    for m in measurements:
        df = pd.DataFrame(m['measurements'], columns=m['columns'])
        new_max_value = float(df.max().value)
        if new_max_value > max_value:
            max_value = new_max_value
    return max_value

def get_dimension_values(dimensions, dimension_names):
    '''Return the dimension values as a dot delimited string using the dimension_names as a selector'''
    values = ''
    for dim in dimension_names:
        if values != '':
            values += '.'
        values += dimensions[dim]
    return values

def get_all_dimension_values(dimensions):
    '''Return all the dimension values'''
    return '.'.join(dimensions.values())

def plot_measurements(measurements, dimension_names):
    num_rows = len(measurements)

    if num_rows < 1:
        return

    fig, axes = plt.subplots(nrows=num_rows)

    row = 0
    fig.suptitle(measurements[0]['name'])

    max_value = get_max_value(measurements)

    for r in measurements:
         df = pd.DataFrame(r['measurements'], columns=r['columns'])
         ts = pd.Series(data=df.value.tolist(), index=[pd.tseries.tools.to_datetime(t) for t in df.timestamp]) 
         plot = ts.plot(ax=axes[row])
         plot.locator_params(nbins=1,axis='y')
         ax = axes[row]
         ax.yaxis.set_label_position("right")
         ylabel = get_dimension_values(r['dimensions'], dimension_names)
                
         ax.set_ylabel(ylabel , rotation=0, horizontalalignment='right')
         ax.yaxis.set_label_position("right")
         if row != num_rows - 1:
            ax.xaxis.set_ticklabels([])
         ax.set_ylim([0.0, max_value])
         row = row + 1
            
def boxplot_measurements(measurements):
    dfs = []
    for m in measurements:
        df = pd.DataFrame(m['measurements'], columns=m['columns'])
        df['name'] = [get_all_dimension_values(m['dimensions']) for k in xrange(len(m['measurements']))]
        dfs.append(df)

    df = pd.concat(dfs)
    boxplot = df.boxplot(column='value', by='name')
    plt.xticks(rotation=45)
    

    
```

```python
measurements = metrics.list_measurements(**{'name':'net_bytes_in', 'start_time':'2014-04-30T11:00:00Z'})
dfs = measurements_to_dfs(measurements)
plot_measurements(measurements, ['hostname', 'device_name'])
measurements = metrics.list_measurements(**{'name':'cpu_user_perc', 'start_time':'2014-04-30T11:00:00Z'})
dfs = measurements_to_dfs(measurements)
plot_measurements(measurements, ['hostname'])
```

```python
measurements = metrics.list_measurements(**{'name':'cpu_user_perc', 'start_time':'2014-04-30T15:00:00Z'})
boxplot_measurements(measurements)
```
